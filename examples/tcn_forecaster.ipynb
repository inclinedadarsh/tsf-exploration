# TCN Forecaster Example

This notebook is an example of how to use the `TCNForecaster` class to forecast time series data using a TCN based model.

Also note that this should be used just for exploration purpose, and nothing else.

```python
from aeon.datasets import load_airline
y = load_airline()
```

```python
from tsf_exploration.forecasting.deep_learning._tcn import TCNForecaster
```

```python
model = TCNForecaster(
    horizon=4,
    window_size=8,
    nb_filters=128,
    kernel_size=3,
    nb_stacks=1,
    epochs=50
)
```

```python
model.fit(y)
```

```python
y_pred = model.predict(y)
```

## Let's plot it!

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(np.append(y, y_pred))
plt.plot(y)
```
